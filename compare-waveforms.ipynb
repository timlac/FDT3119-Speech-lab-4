```python
import torch
import torchaudio
import numpy as np
import matplotlib.pyplot as plt
```

```python
example = torch.load('lab4_example.pt')
```

```python
example_waveform = example["data"][1][0]
example_sample_rate = example["data"][1][1]

print("Waveform shape:", example_waveform.shape)
print("Sample rate:", example_sample_rate)
```

```python
from IPython.display import Audio

samples = example_waveform
samplerate = example_sample_rate

Audio(samples, rate=samplerate)
```

```python
duration = example_waveform.shape[1] / example_sample_rate
time = np.linspace(0., duration, example_waveform.shape[1])
```

```python
plt.figure(figsize=(10, 4))
plt.plot(time, example_waveform[0, :])
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
mel_spec_transform = torchaudio.transforms.MelSpectrogram(example_sample_rate, n_mels=80)
mel_spectrogram = mel_spec_transform(example_waveform)
```

```python
plt.pcolormesh(mel_spectrogram.numpy()[0], cmap='viridis')
plt.title("My Mel Spectrogram")
plt.show()
```
